# Oriented Core with Plotly

```python
center = (100,200,2100)
radius, height = 2, 5

trd_hole, plg_hole = 275.63, 58.64
alpha, beta, gamma, betaL = 50, 337, 24, 7

stk, dip = 350.102, 69.744   # Planar Structure
trd, plg = 110.502, 66.836   # Linear Structure
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig = go.Figure(data=(
                      cylinder3D (center,radius,-height,trd_hole,plg_hole,(0,255,0),.1) +          # core / drill hole ↑
                      cylinder3D (center,radius, height,trd_hole,plg_hole,(0,255,0),.2) +          # core / drill hole ↓

                      plane3D (center,radius,stk,dip,(255,128,0),.8) +                             # PLANE v1
                      disk3D (center,radius,stk,dip,(255,128,0),1,ddr='y')  +                      # PLANE v2
                      plane3D_symb (center,radius,stk,dip,(0,0,0),num='n') +                       # PLANE / labeling strike n' dip
                      pole3D (center,radius,stk,dip,(0,0,0),hem='s') +                             # PLANE / pole
                      line3D (center,radius,trd,plg,(0,152,70),hem='s') +                          # LINE  / labeling trend n' plunge


                      core_axis (center,height,trd_hole,plg_hole,(155,155,155)) +                  # L CA   → Core Axis
                      bottom_hole (center,radius,height,trd_hole,plg_hole,(0,0,255)) +             # L BOH  → Bottom of hole line
                      long_axis (center,radius,height,trd_hole,plg_hole,beta,(0,0,0)) +            # L E E' → Lower-Upper end of Long axis of Intersection elipse

                      angle_alpha (center,radius,trd_hole,plg_hole,alpha,beta,(255,0,0),hem='n') + # ∡ α  alpha
                      angle_beta (center,radius,height,trd_hole,plg_hole,beta,(0,255,0)) +         # ∡ β  beta
                      angle_betaL (center,radius,height,trd_hole,plg_hole,betaL,(0,255,0)) +       # ∡ βL betaL
                      angle_gamma (center,radius,trd_hole,plg_hole,alpha,beta,gamma,(0,255,0)) +   # ∡ γ  gamma
                      line_gamma (center,radius,trd_hole,plg_hole,alpha,beta,gamma,(0,152,70))
                      ))


fig.add_traces(north3D((center[0]+height-2,center[1]-height+2,center[2]),2))                      # symbol north
fig.add_traces(cube3D(center,(height)*2))                                                         # aspect cube / escalas iguales

fig.update_layout(layout_core ())
#fig.show()

#display(HTML(fig.to_html(include_plotlyjs='cdn')))
#camera_button()                                                                                   # 3D Camera Controls
```

```python
fig = go.Figure()
fig = go.Figure(data=(
                      cylinder3D (center,radius,-height,trd_hole,plg_hole,(0,255,0),.1) +          # core / drill hole ↑
                      cylinder3D (center,radius, height,trd_hole,plg_hole,(0,255,0),.2) +          # core / drill hole ↓
                      plane3D (center,radius,stk,dip,(255,128,0),.8) +                             # PLANE v1
                      #disk3D (center,radius,stk,dip,(255,128,0),1,ddr='y')  +                      # PLANE v2
                      plane3D_symb (center,radius,stk,dip,(0,0,0),num='n') +                       # PLANE / labeling strike n' dip
                      pole3D (center,radius,stk,dip,(0,0,0),hem='s') +                             # PLANE / pole
                      line3D (center,radius,trd,plg,(0,152,70),hem='s') +                          # LINE  / labeling trend n' plunge
                      core_axis (center,height,trd_hole,plg_hole,(155,155,155)) +                  # L CA   → Core Axis
                      bottom_hole (center,radius,height,trd_hole,plg_hole,(0,0,255)) +             # L BOH  → Bottom of hole line
                      long_axis (center,radius,height,trd_hole,plg_hole,beta,(0,0,0)) +            # L E E' → Lower-Upper end of Long axis of Intersection elipse
                      angle_alpha (center,radius,trd_hole,plg_hole,alpha,beta,(255,0,0),hem='n') + # ∡ α  alpha
                      angle_beta (center,radius,height,trd_hole,plg_hole,beta,(0,255,0)) +         # ∡ β  beta
                      angle_betaL (center,radius,height,trd_hole,plg_hole,betaL,(0,255,0)) +       # ∡ βL betaL
                      angle_gamma (center,radius,trd_hole,plg_hole,alpha,beta,gamma,(0,255,0)) +   # ∡ γ  gamma
                      line_gamma (center,radius,trd_hole,plg_hole,alpha,beta,gamma,(0,152,70))
                      ))
#fig.add_traces(north3D((center[0]+height-2,center[1]-height+2,center[2]),2))                      # symbol north
fig.add_traces(cube3D(center,(height)*2))                                                         # aspect cube / escalas iguales


##
# New Layou para modificar la posición de la cámara
camera = dict(up=dict(x=0, y=0, z=1), center=dict(x=0.014764559450277355, y=0.026011701960465337, z=-0.09493583691104597), eye=dict(x=0.7367940534083376, y=-0.0833587255918691, z=0.0065121287692052054))

#fig.update_layout(scene_camera=camera)

fig.layout.annotations = []           # Ocultar anotaciones
fig.update_layout(
    showlegend=False,                 # Ocultar leyenda
    title=None,                       # Ocultar título
    annotations=[],                   # Ocultar anotaciones
    scene=dict(
        xaxis=dict(visible=False),    # Ocultar eje X
        yaxis=dict(visible=False),    # Ocultar eje Y
        zaxis=dict(visible=False),    # Ocultar eje Z
        bgcolor="white",              # Cambiar el fondo a blanco
        camera=camera                 # Configurar la cámara
        ),
    margin=dict(l=1, r=1, t=1, b=1),  # Reducir márgenes
    width=700, height=500             # Establecer ancho / altura de la gráfica
)

# Configuración personalizada para la exportación de imágenes desde Plotly | buttom → 'Download plot as a png'
config = {
    'toImageButtonOptions': {
        'format': 'png',        # Formato de la imagen | 'png' 'jpeg' 'svg' 'webp'
        'filename': 'newplot',  # Nombre del archivo exportado | 'newplot'
        'height': None,         # Altura de la imagen exportada (por defecto usa la altura de la gráfica) | None
        'width': None,          # Ancho de la imagen exportada (por defecto usa la anchura de la gráfica) | None
        'scale': 10             # Factor de escala para mejorar la calidad | 1
    }
}
# Mostrar figura con configuración personalizada
#fig.show(config=config)       # no ejecutado, por que renderiza dos veces, se llama en la siguiente linea
##


# Mostrar gráfico y agregar botón
display(HTML(fig.to_html(include_plotlyjs='cdn', config=config)))
camera_button()
```
